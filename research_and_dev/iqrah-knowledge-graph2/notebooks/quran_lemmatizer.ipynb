```python
import csv

def parse_quranic_corpus(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file, delimiter='\t')
        corpus = list(reader)
    return corpus

corpus = parse_quranic_corpus("../data/quran-morphology-v0.5.csv")
```
